```python
import pandas as pd
```

```python
compound_selection = pd.read_csv('data/1_first_stage_compound.csv')
df = pd.read_csv('data/3_docking_results.csv')
```

```python
filter_list = df[(df['protein'] == 'P42345') & (df['energy'] < -8.5)]['Source_idx'].unique()
```

```python
df1 = df[~df['Source_idx'].isin(filter_list)]
```

```python
def get_top20_mean(group):
    top20 = group.nsmallest(20, 'energy')
    return top20['energy'].mean()

result = df1.groupby('Source_idx').apply(get_top20_mean).reset_index()
result.columns = ['Source_idx', 'top20_mean_energy']
```

```python
result = result.sort_values(by='top20_mean_energy').reset_index(drop=True)
```

```python
top_20_compounds = result.head(20)
```

```python
top_20_compounds['Smiles'] = top_20_compounds['Source_idx'].map(compound_selection.set_index('Source_idx').to_dict()['Smiles'])
```
